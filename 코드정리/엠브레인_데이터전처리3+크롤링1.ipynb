#### 전체(남+여)

```python
gmp_game_ppl = ATG_mp_game.panel_id.unique()
ppl_len = len(gmp_game_ppl)

gmp_game_category = list(ATG_mp_game.category.unique())
## (전체) ATG_mp_game의 category별 Total_UsageTime 평균 / Total_AppCount 평균 / 사용자 게임 이용 비율
time = {}
appcount = {}
user_per = {}
for cat in gmp_game_category:
    df = ATG_mp_game[ATG_mp_game.category == cat]
    time[cat] = round(np.average(df.Total_UsageTime)) 
    appcount[cat] = round(np.average(df.Total_AppCount)) 
    user_per[cat] = round(sum(ATG_mp_game.loc[:,cat]) / ppl_len,2)
    
category = gmp_game_category
times = list(time.values())
appcounts = list(appcount.values())
user_pers = list(user_per.values())
all_game_category = pd.DataFrame({'category':category,'time_avg':times,'appcount_avg':appcounts,'user_percent':user_pers})
all_game_category.sort_values(by='time_avg',ascending=False)
```

## 패널별 게임이용 경험 합치기

```python
ATG_mp_game

atg_mp_game = ATG_mp_game.loc[:,:'age']
atg_mp_game.category = [a.replace(' ','') for a in atg_mp_game.category]

category = list(atg_mp_game.category.unique())
ppl = list(atg_mp_game.panel_id.unique())

cols = {}
h = 0
for p in ppl:
    df = atg_mp_game[atg_mp_game.panel_id == p].reset_index(drop=True)
    df_cat = df.category
    
    if h == 0:
        cols['panel_id'] = []
        cols['panel_id'].append(p)
        for i in range(len(df)):
            dc = df.category[i]
            cols[dc+'_time'] = []
            cols[dc+'_appcount'] = []
            cols[dc+'_time'].append(df.Total_UsageTime[i])
            cols[dc+'_appcount'].append(df.Total_AppCount[i])
    
        df_not_cat = list(set(category) - set(df_cat))
        for n in df_not_cat:
            cols[n+'_time'] = []
            cols[n+'_appcount'] = []
            cols[n+'_time'].append(0)
            cols[n+'_appcount'].append(0)
        
        h += 1
        
    else:
        cols['panel_id'].append(p)
        for i in range(len(df)):
            dc = df.category[i]
            cols[dc+'_time'].append(df.Total_UsageTime[i])
            cols[dc+'_appcount'].append(df.Total_AppCount[i])
    
        df_not_cat = list(set(category) - set(df_cat))
        for n in df_not_cat:
            cols[n+'_time'].append(0)
            cols[n+'_appcount'].append(0)
        
        
ATG_mp_cor = pd.concat([pd.Series(cols[x]) for x in list(cols.keys())],axis = 1)
ATG_mp_cor.columns = list(cols.keys())

aa = atg_mp_game.drop_duplicates(subset='panel_id')
aa = aa.loc[:,['panel_id','gender','age']]
A = pd.merge(aa,ATG_mp_cor,on='panel_id')
A

```

```python
### 각각별로(열별) 0-5점 만들기

A = AA
col = A.columns

for i in range(3,37):
    print(col[i])
    df = A.loc[:,col[i]]
    reset = df.loc[[True if s > 0 else False for s in df]].reset_index(drop=True)
    
    s_1 = round(np.percentile(reset,20))
    s_2 = round(np.percentile(reset,40))
    s_3 = round(np.percentile(reset,60))
    s_4 = round(np.percentile(reset,80))
    s_5 = round(np.percentile(reset,100))
    
    new_score = []
    for j in df:
        if j == 0:
            new_score.append(0)
        elif j <= s_1:
            new_score.append(1)
        elif j <= s_2:
            new_score.append(2)
        elif j <= s_3:
            new_score.append(3)
        elif j <= s_4:
            new_score.append(4)
        elif j <= s_5:
            new_score.append(5)
    
    A.loc[:,col[i]] = new_score
    
A    
    
```
