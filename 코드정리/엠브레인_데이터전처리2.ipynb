```python
# AST1_mp --> AST1_mp_time

AST_D1 = {}
print(len(AST1_mp))
for i in range(len(AST1_mp)):
    try:
        if AST_D1.get(AST1_mp['app_name'][i],0) == 0:
            AST_D1[AST1_mp['app_name'][i]] = AST1_mp['total_time_stamp'][i]
        else:
            AST_D1[AST1_mp['app_name'][i]] += AST1_mp['total_time_stamp'][i]
    except:
        print(AST1_mp['app_name'][i],':',AST1_mp['total_time_stamp'][i])
        continue

AST_D1_name = []
AST_D1_time = []
AST_Keys = list(AST_D1.keys())
for i in range(len(AST_Keys)):
    AST_D1_name.append(AST_Keys[i])
    AST_D1_time.append(AST_D1[AST_Keys[i]])
AST1_mp_time = pd.DataFrame({'App_name':AST_D1_name,'Time':AST_D1_time})
AST1_mp_time = AST1_mp_time.sort_values(by='Time', ascending = False)
AST1_mp_time = AST1_mp_time.reset_index(drop=True)

AST1_mp_time['week_avg_hour'] = [int((i/4)/(60*60)) for i in AST1_mp_time['Time']]
AST1_mp_time['day_avg_ho'] = [int(i/7) for i in AST1_mp_time['week_avg_hour']]

AST1_app_id = AST1_mp.drop_duplicates(subset=['panel_id','app_name'])
AST1_app_id = AST1_app_id.iloc[:,[0,3]]

AST1_peo = {}
for name in AST1_app_id['app_name']:
    if AST1_peo.get(name,0) == 0:
        AST1_peo[name] = 1
    else:
        AST1_peo[name] += 1
        
AST1_mp_time['user_sum'] = [AST1_peo[name] for name in AST1_mp_time['App_name']]

AST1_per = []
sum_p = len(AST1_mp.panel_id.unique())
for i in AST1_mp_time['user_sum']:
    AST1_per.append(round(i/sum_p,2))
AST1_mp_time['user_percent'] = AST1_per

AST1_user_day_ho = []
day = AST1_mp_time['day_avg_ho']
user = AST1_mp_time['user_sum']
for i in range(len(AST1_mp_time)):
    AST1_user_day_ho.append(round(day[i]/user[i],2))
AST1_mp_time['day_user_avg_ho'] = AST1_user_day_ho
AST1_mp_time
```

## GAME_Data

### 1) ATM_mp_game
#### : 앱 분류가 ATG에 비해 제대로 되어있지 않아 만들기만 하고 이용은 ATG_mp_game으로

```python
# ATM_mp에서 sub_category가 게임인 경우만
# ATM의 sub_category:게임을 기준으로 했더니 대응하는 category가 이상해져버림
# 게임만을 하려면 ATG를 이용하는게 나음
ATM_mp_game_df = ATM_mp[ATM_mp['sub_category']== '게임']

M_game_id = ATM_mp_game_df.panel_id.unique()
print(len(M_game_id))
h = 1
j = 1

for mid in M_game_id:
    print(j)
    df = ATM_mp_game_df.loc[ATM_mp_game_df['panel_id']== mid,:].reset_index(drop=True)
    df_cate = {}
    df_cate_appcount = {}
    for i in range(len(df)):
        if df_cate.get(df['category'][i],0) == 0:
            df_cate[df['category'][i]] = df['Total_UsageTime'][i]
            df_cate_appcount[df['category'][i]] = df['Total_AppCount'][i]
             
        else:
            df_cate[df['category'][i]] += df['Total_UsageTime'][i]
            df_cate_appcount[df['category'][i]] += df['Total_AppCount'][i]
             
                    
    cate_cate = list(df_cate.keys())
    cate_time = list(df_cate.values())
    cate_app = list(df_cate_appcount.values())
    cate_length = len(cate_cate)

    age = int(APP_pd.loc[APP_pd['panel_id'] == mid,'age'].values)
    gender = int(APP_pd.loc[APP_pd['panel_id'] == mid,'gender'].values)
    
    if h == 1:
        a = pd.DataFrame({'panel_id':[mid]*cate_length,'category':cate_cate,'Total_UsageTime':cate_time,
                          'Total_AppCount': cate_app,'gender':[gender]*cate_length,'age':[age]*cate_length})
        h = 2
        j += 1
    else:
        a2 = pd.DataFrame({'panel_id':[mid]*cate_length,'category':cate_cate,'Total_UsageTime':cate_time,
                          'Total_AppCount': cate_app,'gender':[gender]*cate_length,'age':[age]*cate_length})
        a = pd.concat([a,a2],axis = 0)
        j += 1
        

ATM_mp_game = a
ATM_mp_game
```

```python
ATM_mp_game.to_csv('ATM_mp_game.csv',index=False)
```
