```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
os.chdir('E:\\엠브레인')
md = pd.read_csv('mpd.csv')
female = md[md.X0001==2]
male = md[md.X0001==1]
print(len(female), len(male))
teen = md[md.X0002==1]
twen = md[md.X0002==2]
thir = md[md.X0002==3]
four = md[md.X0002==4]
fif = md[md.X0002==5]
six = md[md.X0002==6]
print(len(teen), len(twen), len(thir), len(four), len(fif), len(six))
fm10 = female[female.X0002==1]
fm20 = female[female.X0002==2]
fm30 = female[female.X0002==3]
fm40 = female[female.X0002==4]
fm50 = female[female.X0002==5]
fm60 = female[female.X0002==6]
m10 = male[male.X0002==1]
m20 = male[male.X0002==2]
m30 = male[male.X0002==3]
m40 = male[male.X0002==4]
m50 = male[male.X0002==5]
m60 = male[male.X0002==6]
print(len(fm10), len(fm20), len(fm30), len(fm40), len(fm50), len(fm60))
print(len(m10), len(m20), len(m30), len(m40), len(m50), len(m60))
```

```python
#모바일게임하는 여자
mobile_fm = female[female['K0002']==1]
mobile_cat = ['K0034', 'K_action', 'K0037', 'K0038', 'K0039', 'K0040', 'K0041', 'K0042']
mobile_name = ['롤플레잉','액션', '스포츠', '전략시뮬', '퍼즐/보드', '아케이드', '캐주얼 게임', '카지노']
print(len(mobile_cat), len(mobile_name))
mobile_fm_num = len(mobile_fm)
print(mobile_fm_num)

#게임 카테고리 별 경험 비율
mobile_fm_df = {}
for cat in mobile_cat:
    print(cat, ':', round(sum(list(mobile_fm[cat]))/mobile_fm_num*100), '%')
    mobile_fm_df[cat] = round(sum(list(mobile_fm[cat]))/mobile_fm_num, 2)*100

genre = list(mobile_fm_df.keys())
fm = list(mobile_fm_df.values())
fm_mobile = pd.DataFrame({'mobile_genre': genre, 'female_percent':fm})
fm_mobile['mobile_genre'] = mobile_name
fm_mobile
```
