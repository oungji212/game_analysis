```python
feat2 = ['액션_time', '스포츠_time','퍼즐_time', '캐주얼게임_time',
         '단어_time', '교육_time', '퀴즈_time', '롤플레잉_time', '아케이드_time',
         '전략_time', '어드벤처_time','카드_time', '시뮬레이션_time', '보드_time',
         '자동차경주_time', '음악_time','카지노_time']
```

```python
def clustering_anal(df,feat):
    cols = feat
    for col in cols:
        print(col)
        features = list(set(df[col]))
        features.sort()
        for feature in features:
            num = df[col].value_counts()[feature]
            print(feature,':',num,'/',round(num/len(df),2)*100,'%')
        print('------------------------------------------------------')
```

```python
def playing_other_games(df,feat2,removes):
    cols = feat2.copy()
    cols.remove(removes+'_time')
    genre = []
    value = [] 
    for col in cols:
        genre.append(col.replace('_time',''))
        value.append(sum([1 if df[col][i] > 0 else 0 for i in range(len(df))]) / len(df))
    plot_df = pd.DataFrame({'genre':genre,'value':value})
    sns.barplot(x ='genre',y='value',data= plot_df)
    plt.xticks(rotation=90)
    plt.show()
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```
