### 최고매출 (profit)

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
import urllib.request

driver = webdriver.Chrome()

# 주요 모바일 게임업계들이 낸 게임들도 찾아봐야함

# 최고 매출 
url = 'https://play.google.com/store/apps/category/GAME/collection/topgrossing'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html,'lxml')
url_list = soup.select('div.details a.title')
urls = [] # 
for ur in url_list:
    game_site = ur.get('href')
    urls.append('https://play.google.com'+game_site)

rank = []
game_name = []
game_company = []
game_category = []
game_country = []
game_star = []
game_download = []


for i in range(len(urls)):
    rank.append(i+1)
    print(i)
    
    driver.get(urls[i])
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    
    name = soup.select('div.sIskre h1 span')
    name = name[0].text
    print(name)
    game_name.append(name)
    
    company = soup.select('div.qQKdcc span a')
    company = company[0].text
    print(company)
    game_company.append(company)
    
    category = soup.select('span a.hrTbp.R8zArc')
    category = category[1].text
    print(category)
    game_category.append(category)
    
    star = soup.select('div.K9wGie div.BHMmbe')
    star = star[0].text
    print(star)
    game_star.append(star)
    
    down = soup.select('div.IQ1z0d span.htlgb')
    down = down[4].text
    print(down)
    game_download.append(down)
    
    country = soup.select('div.IQ1z0d span.htlgb div')
    country = country[-1].text
    print(country)
    game_country.append(country)
    
    
driver.close
```

```python
import pandas as pd
import numpy as np

# rank 
# game_name 
# game_company 
# game_category 
# game_star 
# game_download 

game_profit = pd.DataFrame({'rank':rank,'name':game_name,'company':game_company,'category':game_category,'star':game_star,'download':game_download,'country':game_country})
game_profit
```

```python
country2 = []
country_list = game_profit.country

for con in country_list:
    con = con.lower()
    if 'korea' in con or 'seoul' in con:
        country2.append('korea')
    elif '개인정보' in con:
        country2.append('No information')
    elif '京' in con:
        country2.append('other country')
    elif '06141' == con:
        country2.append('korea')
    else:
        i = 1
        for s in con:
            if ord(s) <= ord('z'):
                if i == len(con):
                    country2.append('other country')
                else:
                    i += 1
                    continue
            elif ord(s) > ord('z'):
                country2.append('korea')
                break
    
country2
```

```python
game_profit['country2'] = country2
game_profit
```

```python
driver = webdriver.Chrome()
url = 'http://www.gevolution.co.kr/'
driver.get(url)

driver.find_element_by_xpath('//*[@id="btn_loginLayer"]').click()
driver.find_element_by_xpath('//*[@id="loginSec"]/div[3]/a/img').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="naverSec"]/div[1]/fieldset/div/a/img').click()
time.sleep(1)
ID = 'oungji212'
PW = 'tpfpql2!'
driver.execute_script("document.getElementsByName('id')[0].value=\'" + ID + "\'")
driver.execute_script("document.getElementsByName('pw')[0].value=\'" + PW + "\'")
driver.find_element_by_xpath('//*[@id="frmNIDLogin"]/fieldset/input').click()

driver.find_element_by_xpath('//*[@id="GamesMainList"]/div/div/div[2]/h3/span/a').click()
driver.get('http://www.gevolution.co.kr/rank/aos_us.asp')

# 구글 플레이스토어 미국 (매출)

game_rank = []
game_name = []
game_company = []
game_genre = []
game_download = []

for i in range(1,61):
    print(i)
    path = '//*[@id="imgload"]/table/tbody/tr[{}]/td[4]/div/div/span[1]/a'.format(i)
    driver.find_element_by_xpath(path).click()
    time.sleep(10)
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    
    game_rank.append(i)
    name = soup.select('div#gameArea div#local_navi i')
    name = name[0].text
    print(name)
    game_name.append(name)
    
    company = soup.select('dl.svc_info dd span a')
    company = company[0].text
    company = company.replace(' [게임 보기]','')
    print(company)
    game_company.append(company)
    
    gendown = soup.select('div.rankstl ol li span.rank span.rankText')
    print(gendown)
    
    try:
        span = []
        for txt in gendown:
            if len(txt.text) >=2 :
                span.append(txt)
            else:
                continue
    
        genre = span[0].text
    except:
        genre = 'Search'
    
    print(genre)
    game_genre.append(genre)
    
    driver.back()

us_google_profit = pd.DataFrame({'rank':game_rank,'name':game_name,'company':game_company,'genre':game_genre})
```
